```python
#continuous - histogram , boxplot ,2var- scatter plot , lineplot discrete - bar plot , pie chart , stacked bar , un stacked bar 
```

```python
plt.hist(df['total_bill'], bins = 10)
plt.xlabel("total bill")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(df['total_bill'])
plt.show()
```

```python
plt.scatter(df['total_bill'] , df['tip'])
plt.xlabel("Bill")
plt.ylabel("Tip given")
plt.show()
```

```python
company = ['google' , 'amazon' , 'microsoft']
revenue = [15 , 18 , 26]
```

```python
plt.bar(company , revenue)
plt.show()
```

```python
plt.pie(revenue,labels=company,autopct ="%0.1F%%")
plt.show()
```

```python
#seaborn - histogram , boxplot , scatter plot , lineplot , violine plot , joint plot , pair plot , heat map
```

```python
sns.histplot(df['total_bill'] ,kde=True , bins = 10)
plt.show()
```

```python
sns.kdeplot(df['total_bill'] , fill = True)
plt.show()
```

```python
sns.boxplot(df['total_bill'])
plt.show()
```

```python
sns.scatterplot(x=df['total_bill'] , y =df['tip'],hue=df['sex'])
plt.show()
```

```python
sns.relplot(x=df['total_bill'] , y = df['tip'], hue=df['sex'])
plt.show()
```

```python
sns.jointplot(x = df['total_bill'] , y=df['tip'])
plt.show()
```

```python
sns.violinplot(x=df['total_bill'] ,y =df['sex'])
plt.show()
```

```python
sns.pairplot(df,vars = ['total_bill' , 'tip'])
plt.show()
```

```python
c = df[['total_bill' , 'tip']].corr()
```

```python
sns.heatmap(c , annot = True)
plt.show()
```

```python
sns.countplot(x=df['time'])
plt.show()
```
